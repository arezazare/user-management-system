```python
import logging
from auth import login
from account_manager import AccountManager
from account_storage import seed_admin, save_accounts

# Configure logging if needed
logging.basicConfig(
    filename="user_activity.log",
    level=logging.INFO,
    format="%(asctime)s:%(levelname)s:%(message)s",
    datefmt="%m/%d/%Y %I:%M:%S %p"
)

if __name__ == "__main__":
    # Seed the admin account (if one does not already exist)
    seed_admin()
    # Prompt the user to log in or create a new account
    manager = AccountManager()
    
    account_login = input("Do you have an account? (y/n): ").strip().lower()
    
    if account_login == "y":

        user = login() # Use the login from auth.py
        
        if user:
            print("âœ… Logged in successfully.")
            logging.info(f"User {user['username']} has logged in.")
            # Call post_login_menu after successful login
            manager.post_login_menu(user)
        else:
            logging.warning("User login failed.")
            print("âŒ Login failed. Please try again.")
            
    elif account_login == "n":
        print("ğŸ” Creating a new account...")
        profile = manager.account_creation()
        save_accounts(profile)
    else:
        logging.warning("Invalid input for account login/creation.")
        print("âŒ Invalid input. Please enter 'y' or 'n'.")
```
